```python
#merge the dataframes
from functools import reduce

# List of dataframes to merge
dataframes = [passing_stats, shot_creation_actions, shooting_stats, pass_completion, standard_stats, Player_possesion, Playing_time]

for df in dataframes:
    df.columns = df.columns.str.lower()
# Merge all dataframes on the common columns
merged_df = reduce(lambda left, right: pd.merge(left, right, on=['player'], how='outer', suffixes=('', '_dup')), dataframes)

# Remove duplicated columns
merged_df.columns = merged_df.columns.astype(str)
merged_df = merged_df.loc[:, ~merged_df.columns.str.endswith('_dup')]


# Print the columns of the merged dataframe to verify
print(merged_df.columns)
```

```python
#Long passes data
# Importing libraries
import matplotlib.pyplot as plt
import seaborn as sns
from adjustText import adjust_text

# Define Liverpool midfielders
Liverpool_Offensive_Midfielders = {'Dominik Szoboszlai', 'Ryan Gravenberch', 'Alexis Mac Allister', 'Curtis Jones', 'Harvey Elliott'}

# Filter players who have played at least 10 full 90s
filtered_data = Relevant_data[Relevant_data['no._of_90s_played'] > 10]

# Calculate percentiles for outlier detection
low_cutoff = filtered_data['long_passes_attempted'].quantile(0.3)  # Bottom 30% cutoff
high_cutoff = filtered_data['long_passes_attempted'].quantile(0.85)  # Top 15% cutoff

# Set theme
sns.set_theme(style='dark')

# Create figure
fig, ax = plt.subplots(figsize=(12, 7), facecolor='black')
ax.set_facecolor((0.1, 0.4, 0.2))

# Scatter plot for all players
sns.scatterplot(data=filtered_data, x='long_pass_completion_percentage', y='long_passes_attempted', 
                color='yellow', edgecolor='black', alpha=0.7)

# Scatter plot for Liverpool players (highlighted in red)
liverpool_data = filtered_data[filtered_data['player'].isin(Liverpool_Offensive_Midfielders)]
sns.scatterplot(data=liverpool_data, 
                x='long_pass_completion_percentage', y='long_passes_attempted', 
                color='red', edgecolor='black', s=100, label="Liverpool Players")

# Title and labels
plt.title('Long Pass Completion Rate\nNo. of 90s played >10', 
          fontsize=14, fontweight='bold', color='white', fontfamily='serif')
plt.xlabel('Long Passes Completed %', fontsize=12, color='white', fontfamily='serif')
plt.ylabel('Long Passes Attempted', fontsize=12, color='white', fontfamily='serif')

# Change tick colors to white for better visibility
ax.tick_params(colors='white')

# Annotate Liverpool players + middle/top outliers
texts = []
for i, row in filtered_data.iterrows():
    if row['player'] in Liverpool_Offensive_Midfielders or row['long_passes_attempted'] >= high_cutoff:
        color = 'black' if row['player'] in Liverpool_Offensive_Midfielders else 'white'
        text = plt.annotate(row['player'],
                            (row['long_pass_completion_percentage'], row['long_passes_attempted']), 
                            fontsize=8, ha='right', color=color, va='bottom', fontfamily='serif')
        texts.append(text)

# Adjust text positions to avoid overlap
adjust_text(texts, arrowprops=dict(arrowstyle='-', lw=0.5, color='white'))

# Add small "Data from FBref" at the bottom
plt.figtext(0.15, 0.02, "Data from FBref", fontsize=8, color='white', fontfamily='serif')

# Show plot
plt.legend()
plt.show()
```
