```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from adjustText import adjust_text

# Define Liverpool midfielders
Liverpool_Offensive_Midfielders = {
    'Dominik Szoboszlai', 'Ryan Gravenberch',
    'Alexis Mac Allister', 'Curtis Jones', 'Harvey Elliott'
}

# Set style
sns.set_theme(style='white')

# Filter players with at least 3 goals or 3 assists
filtered_data = Relevant_data[
    (Relevant_data['goals'] >= 3) | (Relevant_data['assists'] >= 3)
]

# Create figure
fig, ax = plt.subplots(figsize=(18, 9), facecolor='white')
ax.set_facecolor('lightgray')

# Scatter plot for all filtered players
sns.scatterplot(
    data=filtered_data,
    x='assists', y='goals',
    color='blue', edgecolor='black', alpha=0.7
)

# Scatter plot for Liverpool players (highlighted in bold red)
sns.scatterplot(
    data=filtered_data[filtered_data['player'].isin(Liverpool_Offensive_Midfielders)],
    x='assists', y='goals',
    color='red', edgecolor='black', s=100, label='Liverpool Players'
)

# Titles and labels
plt.title(
    r"Top Performing 'Attacking' Midfielders: Assists X Goals",
    fontsize=14, fontweight='bold', color='black', fontfamily='serif'
)
plt.xlabel('Assists', fontsize=12, color='black', fontfamily='serif')
plt.ylabel('Goals', fontsize=12, color='black', fontfamily='serif')

# Change tick colors to black (you had 'white', but the background is lightgray now)
ax.tick_params(colors='black')

# --- Jitter Setup ---
# Create small random jitter arrays the same length as filtered_data
np.random.seed(42)  # for reproducibility, optional
jitter_x = np.random.uniform(-0.15, 0.15, len(filtered_data))
jitter_y = np.random.uniform(-0.15, 0.15, len(filtered_data))

# Annotate only the filtered players with better spacing
texts = []
for i, row in enumerate(filtered_data.itertuples()):
    # If the row is a namedtuple from iterrows or itertuples, 
    # access columns with row.assists, row.goals, row.player, etc.
    color = 'red' if row.player in Liverpool_Offensive_Midfielders else 'black'
    weight = 'bold' if row.player in Liverpool_Offensive_Midfielders else 'normal'
    
    # Manual nudging for certain players
    x_offset = 0.1 if row.player in Liverpool_Offensive_Midfielders else 0
    y_offset = 0.1 if row.player in [
        'Matheus Cunha', 'Emile Smith Rowe', 'Mateo Kovačić', 'Kevin De Bruyne'
    ] else 0
    
    # Combine manual offsets + random jitter
    x_coord = row.assists + x_offset + jitter_x[i]
    y_coord = row.goals + y_offset + jitter_y[i]
    
    text = plt.annotate(
        row.player,
        (x_coord, y_coord),
        fontsize=11,
        color=color,
        fontweight=weight,
        fontfamily='serif'
    )
    texts.append(text)

# Adjust text positions to reduce overlap
adjust_text(
    texts, ax=ax,
    arrowprops=dict(arrowstyle='-', lw=0.5, color='white', shrinkA=5, shrinkB=5),
    force_points=(1, 2),
    expand_points=(2, 3),
    force_text=(0.5, 1.0),       # Increase if you need more "push" for text
    expand_text=(1.2, 1.2)       # Increase to spread labels further apart
)

# Add small "Data from FBref" at the bottom
plt.figtext(0.15, 0.02, "Data from FBref", fontsize=8, color='white', fontfamily='serif')

# Fix the legend for better visibility
legend = plt.legend(frameon=True, edgecolor="white", facecolor='white', fontsize=10)
legend.get_frame().set_alpha(0.8)  # Slight transparency

# Show plot
plt.show()
```

```python
#We are adding more press resistance data
#Grabbing the names of the 'midfielders' we are interested in

midfielders = midfielders_data2['Player']
press_resistance_table = pd.read_excel(r'C:\Users\ADMIN\Documents\Press resistance table.xlsx')
press_resistance_table = press_resistance_table[press_resistance_table['Player'].isin(midfielders)]
#Add more columns before we create our index
#passing_accuracy, progressive carries,
#passing_accuracy and progressive_carries were both in filtered data
press_resistance_data = press_resistance_data.rename(columns = {'Player':'player'})

pass_and_prog = Relevant_data.loc[
    Relevant_data['player'].isin(midfielders),
    ['player', 'pass_completion_percentage', 'progressive_carries']
]
press_resistance_table.rename(columns={'Player': 'player'}, inplace=True)
final_press_resistance_data = pd.merge(press_resistance_table, pass_and_prog, on='player', how='inner')
final_press_resistance_data['Successful take ons per 90'] = final_press_resistance_data['Sucessful take ons'] / final_press_resistance_data['90s']
final_press_resistance_data['Take ons attempted per 90'] = final_press_resistance_data['Take ons attempted'] / final_press_resistance_data['90s']
final_press_resistance_data['Dispossesed per 90'] = final_press_resistance_data['Dispossesed'] / final_press_resistance_data['90s']
final_press_resistance_data['Miscontrols per 90'] = final_press_resistance_data['Miscontrols'] / final_press_resistance_data['90s']
final_press_resistance_data['Tackled during take ons per 90'] = final_press_resistance_data['Tackled during take ons'] / final_press_resistance_data['90s']
final_press_resistance_data['Successful take on percentage'] = final_press_resistance_data['Successful take ons per 90'] / final_press_resistance_data['Take ons attempted per 90']
final_press_resistance_data['Tackled during take ons percentage'] = final_press_resistance_data['Tackled during take ons per 90'] / final_press_resistance_data['Take ons attempted per 90']
print(final_press_resistance_data.columns)
```

```python
#creating our index
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
df = final_press_resistance_data.copy()
scaler = MinMaxScaler()
weights = {
    'Take ons attempted per 90': 0.20,
    'Succesful take on percentage': 0.45,
    'pass_completion_percentage': 0.20,
    'progressive_carries per 90': 0.20,
    'Tackled during take ons percentage': 0.45,
    'Miscontrols per 90': 0.35,
    'Dispossesed per 90': 0.40
}
for col in ['Succesful take on percentage', 'pass_completion_percentage', 'progressive_carries']:
    df[col + '_norm'] = scaler.fit_transform(df[[col]])
for col in ['Tackled during take ons percentage', 'Miscontrols', 'Dispossesed']:
    df[col + '_norm'] = 1 - scaler.fit_transform(df[[col]])
df['press_resistance_index'] = (
    df['Take ons attempted per 90'] * weights['Take ons attempted per 90'] +
    df['Succesful take on percentage_norm'] * weights['Succesful take on percentage'] +
    df['pass_completion_percentage_norm'] * weights['pass_completion_percentage'] +
    df['progressive_carries_norm'] * weights['progressive_carries per 90'] +
    df['Tackled during take ons percentage_norm'] * weights['Tackled during take ons percentage'] +
    df['Miscontrols_norm'] * weights['Miscontrols per 90'] +
    df['Dispossesed_norm'] * weights['Dispossesed per 90']
)
print(df[['player', 'press_resistance_index']].sort_values(by='press_resistance_index', ascending=False).head(10))
print(df.columns)
```
